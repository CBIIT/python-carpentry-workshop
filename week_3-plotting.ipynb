## Plotting

### `matplotlib` is the most widely used plotting library in Python

In particular, the `pyplot` sub-module of `matplotlib` is the main workhorse.

```python
import matplotlib.pyplot as plt
```

It's pretty simple to use for basic plotting.

```python
time = [0, 1, 2, 3]
position = [-100, 0, 100, 200]
plt.plot(time, position)
```

You can add labels on the axes too.

```python
plt.plot(time, position)
plt.xlabel('Time (hr)')
plt.ylabel('Position (km)')
```

You can change the plot style using a format string.

```python
plt.plot(time, position, 'r--*')
plt.xlabel('Time (hr)')
plt.ylabel('Position (km)')
```

You don't need to display a plot on the screen even though you've created one. E.g., you can create a plot and save it to a file without ever displaying it.

Notebook environments such as Jupyter notebooks automatically *display* plots when they're created.

Other contexts, such as Python scripts and interactive Python sessions, do not automatically display plots when they're created. For this you need to use `plt.show()`.

### Plot data using matplotlib or, if Pandas is being used, Pandas

Plotting from Pandas employs matplotlib in the backend.

Load the data.

```python
# Import relevant library
import pandas as pd

# Define the datafile containing our data of interest
datafile = 'https://raw.githubusercontent.com/swcarpentry/python-novice-gapminder/gh-pages/data/gapminder_gdp_oceania.csv'

# Load the data into a Pandas dataframe
df = pd.read_csv(datafile, index_col='country')
```

Reformat the column labels by stripping off the "gdpPercap_" prefix and optionally convert them to integers.

```python
df.columns = df.columns.str.lstrip('gdpPercap_').astype(int)  # Note the ".astype(int)" is not needed
```

Plot Australia's GPD using Pandas - method 1.

```python
df.loc['Australia', :].plot()
```

Plot Australia's GPD using Pandas - method 2.

```python
df.T['Australia'].plot()
```

Plot Australia's GPD using matplotlib.

```python
years = df.columns
gdp_australia = df.loc['Australia', :]
plt.plot(years, gdp_australia)
```

Plot series of data using Pandas.

```python
df.T.plot.line()  # We could have just used df.T.plot()
```

Plot series of data using matplotlib.

```python
years = df.columns
gdp_australia = df.loc['Australia', :]
gdp_nz = df.loc['New Zealand', :]
plt.plot(years, gdp_australia, '-', label='Australia', color='tab:blue')
plt.plot(years, gdp_nz, '-', label='New Zealand', color='tab:orange')
plt.legend(loc='upper left', title='country')
```

### Many plot types are available

```python
plt.scatter(gdp_australia, gdp_nz)
```

```python
df.T.plot.scatter(x='Australia', y='New Zealand')
```

```python
plt.bar(years, gdp_australia)
plt.bar(years + 0.75, gdp_nz)
```

```python
df.T.plot(kind='bar')
```

### Many plot styles are available

```python
plt.style.use('ggplot')
```

```python
df.T.plot(kind='bar')
```

Go back to the default style.

```python
plt.style.use('default')
```

```python
df.T.plot(kind='bar')
```

### Save your plot to a file using `fig.savefig()`

This way to plot utilizes the object-oriented API (instead of the `pyplot` API), in which the only call to the `plt` module is `subplots()`.

This enables an intuitive way to save figures as well.

```python
fig, ax = plt.subplots()
ax.plot(years, gdp_australia, '-', label='Australia', color='tab:blue')
ax.plot(years, gdp_nz, '-', label='New Zealand', color='tab:orange')
ax.legend(loc='upper left', title='country')
fig.savefig('my_line_plot_using_matplotlib.png')
```

```python
fig, ax = plt.subplots()
df.T.plot.line(ax=ax)
fig.savefig('my_line_plot_using_pandas.png')
```
